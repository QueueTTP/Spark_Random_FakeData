```python
from pyspark.sql import SparkSession
import pandas as pd
import pymysql
```

```python
# 读取第一个 SQL 文件
with open('/Users/qian/Desktop/ZipCode/Python/Spark_Random_FakeData/data_000.sql', 'r') as file:
    sql_query1 = file.read()

# 读取第二个 SQL 文件
with open('/Users/qian/Desktop/ZipCode/Python/Spark_Random_FakeData/data_001.sql', 'r') as file:
    sql_query2 = file.read()

# 创建 MySQL 数据库连接
connection = pymysql.connect(
    host='localhost',
    user='qian',
    password='zipcode0',
    database='millsongs'
)

# 创建一个游标对象
cursor = connection.cursor()

# 将 SQL 文件内容按分号分割为单个语句
sql_statements1 = sql_query1.split(';')
sql_statements2 = sql_query2.split(';')

# 逐条执行 SQL 语句
for statement in sql_statements1:
    if statement.strip():  # 忽略空语句
        cursor.execute(statement)

for statement in sql_statements2:
    if statement.strip():  # 忽略空语句
        cursor.execute(statement)

# 提交更改到数据库
connection.commit()

# 关闭游标和数据库连接
cursor.close()
connection.close()
```
